```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/student-depression-dataset/Student Depression Dataset.csv")
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr_results = df[["Academic Pressure","CGPA","Study Satisfaction","Depression","Work/Study Hours"]].corr()
sns.heatmap(corr_results,annot=True,cmap="cubehelix_r")
plt.show()
```

```python
print(df['Have you ever had suicidal thoughts ?'].unique())
print(df['Family History of Mental Illness'].unique())
print(df['Have you ever had suicidal thoughts ?'].isnull().sum())
print(df['Family History of Mental Illness'].isnull().sum())
```

```python
map ={"Yes":1,"No":0}
df['Have you ever had suicidal thoughts ?'] = df['Have you ever had suicidal thoughts ?'].map(map)
df['Family History of Mental Illness'] = df['Family History of Mental Illness'].map(map)
df
```

```python
corr_results = df[["Academic Pressure","CGPA","Study Satisfaction","Depression","Work/Study Hours",'Have you ever had suicidal thoughts ?','Family History of Mental Illness']].corr()
sns.heatmap(corr_results,annot=True,cmap="cubehelix_r")
plt.show()
```

```python
total_by_gender = df.groupby("Gender")["Depression"].count()
depression_by_gender = df.groupby("Gender")["Depression"].sum()

depression_rate_by_gender = (depression_by_gender / total_by_gender) * 100
print(depression_rate_by_gender)
```

```python
total_by_gender = df.groupby("Gender")["Academic Pressure"].count()
pressure_by_gender = df.groupby("Gender")["Academic Pressure"].sum()

depression_rate_by_gender = (pressure_by_gender / total_by_gender) * 100
print(depression_rate_by_gender)
```

```python
total_by_gender = df.groupby("Gender")["Academic Pressure"].count()
thoughts_by_gender = df.groupby("Gender")['Have you ever had suicidal thoughts ?'].sum()

depression_rate_by_gender = (thoughts_by_gender / total_by_gender) * 100
print(depression_rate_by_gender)
```

We see very similar values ​​in all 3 comparisons. While we understand that the gender difference is not significant, the fact that women have fewer suicidal thoughts in response to high pressure shows that they are slightly more resistant to pressure.
