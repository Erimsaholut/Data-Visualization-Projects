```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/student-lifestyle-dataset/student_lifestyle_dataset.csv")
```

```python
df
```

Let's check the heatmap of this data to interpret it better !

```python
import seaborn as sb
import matplotlib.pyplot as plt

dataplot = sb.heatmap(df.iloc[:,1:].corr(numeric_only=True), cmap="YlGnBu", annot=True)
#we don't need to chech correlations of the studentID


plt.show()
```

We cannot see stress levels, which is an important parameter, in this table. Let's fix that !

```python
levels={
    "Low":0,
"Moderate":1,
"High":2
}
df["Stress_Level"] = df["Stress_Level"].map(levels)
```

```python
dataplot = sb.heatmap(df.iloc[:,1:].corr(numeric_only=True), cmap="YlGnBu", annot=True)

plt.show()
```

When we examine the graph, we see that the highest correlations are between study hours and GPA and study hours and stress level.

This means studying more will definitely increase your GPA, but it will also increase your stress level.

Additionally, we are able to see that physical activities and sleep hours have a reverse correlation, which means they might help to reduce stress. (But donâ€™t forget, they have a weak relationship, around 0.20-0.30.)

```python
df.plot.scatter(x="Study_Hours_Per_Day", y="GPA", title="Study Hours vs. GPA")
df.plot.scatter(x="Study_Hours_Per_Day", y="Stress_Level", title="Study Hours vs. Stress Level")
plt.show()
```
